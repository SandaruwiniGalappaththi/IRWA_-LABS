```python
# Python 3 program to count the uppercase,
# lowercase, special characters
# and numeric values

# Function to count uppercase, lowercase,
# special characters and numbers
def Count(str):
	upper, lower, number, special = 0, 0, 0, 0
	for i in range(len(str)):
		if str[i].isupper():
			upper += 1
		elif str[i].islower():
			lower += 1
		elif str[i].isdigit():
			number += 1
		else:
			special += 1
	print('Upper case letters:', upper)
	print('Lower case letters:', lower)
	print('Number:', number)
	print('Special characters:', special)

# Driver Code
str = "#GeeKs01fOr@gEEks07"
Count(str)

# This code is contributed
# by Sushma Reddy
```

```python
import os
def InvertedIndex():

  dictionary = dict()
  directory = os.getcwd()+'/inverted'
  files = os.listdir(directory)
  
  for file in files:
    with open(os.getcwd()+'/inverted/'+file,'r')as f:
      words = f.read().lower().split()
      
      for word in words:
        if word not in dictionary:
          dictionary[word] =[file]
        else:
          dictionary[word].append(file)
  return(dictionary)

print(dictionary)
```

```python
def and_op(list1,list2):
  if ((list1) and (list2)):
    return set(list1).intersection(list2)
  else:
    return()

ii = InvertedIndex()
for key in InvertedIndex():
  if key == 'schizophrenia':
    list1 = ii[key]
    print('list1',list1)
  if key == 'drug':
    list2 = ii[key]
    print('list2',list2)
print('schizophrenia AND drug',and_op(list1,list2))
```

```python
# import module
import requests

# create a function
# pass the url
def url_ok(url):
	
	# exception block
	try:
	
		# pass the url into
		# request.hear
		response = requests.head(url)
		
		# check the status code
		if response.status_code == 200:
			return True
		else:
			return False
	except requests.ConnectionError as e:
		return e

# driven code
url = "https://www.geeksforgeeks.org/"
url_ok(url)
```
